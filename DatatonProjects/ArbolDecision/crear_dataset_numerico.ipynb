```python
import string

def procesar_columnas(df, columns):
    """
    Preprocesa las columnas de un DataFrame eliminando signos de puntuacin y convirtiendo a minsculas.

    Args:
        df (pd.DataFrame): El DataFrame a procesar.
        columns (list): Lista de nombres de columnas a procesar.

    Returns:
        pd.DataFrame: El DataFrame con las columnas preprocesadas.
    """
    # Crear un diccionario para almacenar los resultados
    for column in columns:
        if column in df.columns:
            # Eliminar signos de puntuacin y convertir a minsculas
            df[column] = df[column].astype(str).apply(
                lambda x: x.translate(str.maketrans('', '', string.punctuation)).lower()
            )
    return df
```

```python
columnas_aprocesar = ['carreraAreaConocimiento','empleoCargoComision','funcionPrincipal','area','puesto']
df_cleaned = procesar_columnas(df_cleaned, columnas_aprocesar)
df_cleaned.head()
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.decomposition import PCA
model_enbeddings = SentenceTransformer('sentence-transformers/LaBSE')
```

```python
import pickle

embeddings_carreras = model_enbeddings.encode(df_cleaned['carreraAreaConocimiento'].tolist())
print(embeddings_carreras)
pca_carreras = PCA(n_components=50)
embeddings_carreras_reducidos = pca_carreras.fit_transform(embeddings_carreras)


# Guardar el modelo PCA para usarlo despus
with open('pca_carreras_50.pkl', 'wb') as f:
    pickle.dump(pca_carreras, f)

df_aux_carreras = pd.DataFrame(
    embeddings_carreras_reducidos,
    columns=[f'carrera_{i+1}' for i in range(50)],
    index=df_cleaned.index  # Alinear ndices
)

df_cleaned = pd.concat([df_cleaned, df_aux_carreras], axis=1)
```
